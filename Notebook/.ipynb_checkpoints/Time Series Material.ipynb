```python
# Time Series Material
import pandas as pd
from pandas import read_csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DatasetTeSelec.csv
DatasetTeSelec = pd.read_csv('datasetTeSelec.csv')
# FeaturesFirstPaper.csv
featuresFirstPaper = pd.read_csv('featuresFirstPaper.csv')
DatasetTeSelec.rename( columns={'Unnamed: 0':'idSerie'}, inplace=True )
featuresFirstPaper.head()
```

```python
# All features
# E: Dataset 1 and 2: CSVs, idSerie: the serie you want to get the value of the features
# O: One plot
def generatePlot_Allfeatures(dataset1, dataset2, idSerie):
    # Get time series 
    dfIdSerie = dataset2.loc[dataset2['idSerie'] == idSerie].reset_index(drop=True)
    # Get dataframe with features
    df_features = dataset1.loc[dataset2['idSerie'] == idSerie].reset_index(drop=True)
    dict_features = df_features.to_dict()
    
    output_str = "Feature Values for the Series\n\n"
    for key, value in dict_features.items():
        output_str += f"{key}: {list(value.values())[0]}\n\n "
    output_str = output_str[:-2]
    
    # Make data for plot for id serie
    dfIdSerie = dfIdSerie.drop(['idSerie'], axis=1).squeeze().dropna().iloc[0:]
    plt.figure(figsize=(18,6))
    plt.grid(linewidth=0.5)
    plt.ylim(2, 3)
    plt.title("Values of time series with id: " + str(idSerie))
    plt.text(0, 4.2, output_str,
        fontsize=10, verticalalignment='top', horizontalalignment='left')
    plt.plot(dfIdSerie.index, dfIdSerie.values)
    # display only every 10th tick label
    plt.xticks(dfIdSerie.index[::20])
    
```

```python
generatePlot_Allfeatures(featuresFirstPaper,DatasetTeSelec, 45)
```

```python
#Get the min and max values for each column in the dataframe
def getMaxandMin(dataset):
    maxSeries = dataset.max()
    minSeries = dataset.min()    
    return maxSeries,minSeries

# E: Dataset 1 and 2: CSVs, feature: the feature you want to get the max and min
# O: Two plots
def generatePlot(dataset1, dataset2, feature):
    
    # Get rows from data set with max and min values
    values = getMaxandMin(dataset1)
    
    minValueData = dataset1.loc[dataset1[feature] == values[1][feature]]
    maxValueData = dataset1.loc[dataset1[feature] == values[0][feature]]
    
    
    # Get time series data from second dataframe transposed
    minValueData = dataset2.loc[dataset2['idSerie'] == minValueData['id'].values[0]].reset_index(drop=True)
    maxTimeSeriesData = dataset2.loc[dataset2['idSerie'] == maxValueData['id'].values[0]].reset_index(drop=True)
   
    # Make data for plot for max value
    maxIdSerie = int(maxTimeSeriesData.iloc[0]['idSerie'])
    maxTimeSeriesData = maxTimeSeriesData.drop(['idSerie'], axis=1).squeeze().dropna().iloc[0:]
   
    plt.figure(figsize=(18,6))
    plt.grid(linewidth=0.5)
    plt.ylim(2, 3)
    plt.title("Max Value of" + feature + " = " + str(values[0][feature]))
    plt.plot(maxTimeSeriesData.index, maxTimeSeriesData.values)
    plt.xticks(maxTimeSeriesData.index[::20],rotation=90)
    # Make data for plot for min value
    minIdSerie = int(minValueData.iloc[0]['idSerie'])
    minValueData = minValueData.drop(['idSerie'], axis=1).squeeze().dropna().iloc[0:]
    print(type(minValueData))
    plt.figure(figsize=(20,6))
    plt.grid(linewidth=0.5)
    plt.ylim(2, 3)
    plt.title("Min Value of " + feature + " = " + str(values[1][feature]))
    plt.plot(minValueData.index, minValueData.values, color='red')
    plt.xticks(minValueData.index[::20],rotation=90)
    
    return 
    
```

```python
generatePlot(featuresFirstPaper,DatasetTeSelec, 'entropy')
```
